```python
from sympy import *

# +x right, +y up, +z forward

def yaw(x):
   return Matrix ([[cos(x), 0, sin(x)],
                  [0, 1, 0],
                  [-sin(x), 0, cos(x)]])

def pitch(x):
   return Matrix ([[1, 0, 0],
                  [0, cos(x), -sin(x)],
                  [0, sin(x), cos(x)]])

def roll(x):
   return Matrix ([[cos(x), -sin(x), 0],
                  [sin(x), cos(x), 0],
                  [0, 0, 1]])


y, p, r = symbols('y p r')
yaw(y) * pitch(p) * roll(0)
```

```python
# +x right, +y up, +z forward

rx, ry, rz, ux, uy, uz, fx, fy, fz = symbols('r_x r_y r_z u_x u_y u_z f_x f_y f_z')
px, py, pz = symbols('p_x p_y p_z')


Matrix([[rx, ry, rz, 0],
        [ux, uy, uz, 0],
        [fx, fy, fz, 0],
        [0, 0, 0, 1]]) * Matrix([[1, 0, 0, px],
        [0, 1, 0, py],
         [0, 0, 1, pz],
         [0, 0, 0, 1]]) 
```
